4.  # Задание 4.

    > Создайте временную таблицу, содержащую 10 лучших студентов 3-го курса института МПСУ
    > (имеющих самую высокую среднюю оценку), окончивших учебу без троек.
    > Отсортируйте ее по убыванию среднего балла.
    > Таблица должна содержать ФИО, номер студенческого билета и средний балл.

    Сначала составим вспомогательный подзапрос `SELECT`:

```python
%%sql
SELECT *
  FROM student
  INNER JOIN students_group
    ON student.students_group_number = students_group.students_group_number
  INNER JOIN field_comprehension
    ON student.student_id = field_comprehension.student_id
  WHERE field_comprehension.mark > 3
    AND students_group.structural_unit_number = 1                           -- MPSU
    AND student.students_group_number ~ '-3'                                -- third year
```

Тк студентов третьего курса, подходящих под условие нет, то попробуем найти студентов 4го курса.

```python
%%sql
SELECT s.surname, s.name, s.patronymic, s.student_id, AVG(field_comprehension.mark) AS "средний балл"
  FROM student AS s
  INNER JOIN students_group
    ON s.students_group_number = students_group.students_group_number
  INNER JOIN field_comprehension
    ON s.student_id = field_comprehension.student_id
  WHERE field_comprehension.mark > 3
    AND students_group.structural_unit_number = 1                           -- MPSU
    AND s.students_group_number ~ '-4'                                      -- fourth year
  GROUP BY s.student_id
  ORDER BY "средний балл" DESC
  LIMIT 10
```

Теперь создадим соответствующую таблицу из подзапроса:

```python
%%sql
CREATE TABLE top_students (
  student_id   INTEGER     PRIMARY KEY,
  surname      VARCHAR(30) NOT NULL,
  name         VARCHAR(30) NOT NULL,
  patronymic   VARCHAR(30) NULL,
  average_mark NUMERIC     NOT NULL,

  CONSTRAINT student_id_key
    FOREIGN KEY(student_id)
    REFERENCES student(student_id)
    ON DELETE CASCADE
)
```

```python
%%sql
INSERT INTO top_students (student_id, surname, name, patronymic, average_mark) (
  SELECT s.student_id, s.surname, s.name, s.patronymic, AVG(field_comprehension.mark) AS average_mark
    FROM student AS s
    INNER JOIN students_group
      ON s.students_group_number = students_group.students_group_number
    INNER JOIN field_comprehension
      ON s.student_id = field_comprehension.student_id
    WHERE field_comprehension.mark > 3
      AND students_group.structural_unit_number = 1                           -- MPSU
      AND s.students_group_number ~ '-4'                                      -- fourth year
    GROUP BY s.student_id
    ORDER BY average_mark DESC
    LIMIT 10
)
```

Убедимся в правильности созданной таблицы:

```python
%%sql
SELECT *
  FROM top_students
```
